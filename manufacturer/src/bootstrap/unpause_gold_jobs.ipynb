```python
from databricks.sdk import WorkspaceClient

w = WorkspaceClient()

gold_job_names = ["gold_supply", "gold_sales", "gold_product"]

jobs = w.jobs.list()
job_ids_to_unpause = []

for job in jobs:
    if job.settings.name in gold_job_names:
        job_ids_to_unpause.append(job.job_id)
        print(f"Found gold job: {job.settings.name} (ID: {job.job_id})")

for job_id in job_ids_to_unpause:
    try:
        job = w.jobs.get(job_id=job_id)
        current_settings = job.settings
        
        if current_settings.trigger and hasattr(current_settings.trigger, 'pause_status'):
            updated_trigger = current_settings.trigger
            updated_trigger.pause_status = "UNPAUSED"
            w.jobs.update(job_id=job_id, new_settings={"trigger": updated_trigger})
            print(f"Unpaused job ID: {job_id}")
    except Exception as e:
        print(f"Error unpausing job {job_id}: {e}")

print("Bootstrap complete - Gold jobs are now unpaused")
```
