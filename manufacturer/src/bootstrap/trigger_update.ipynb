```python
from databricks.sdk import WorkspaceClient
from databricks.sdk.service.jobs import PauseStatus
w = WorkspaceClient()
# change manually
jobs=["gold_product",
      "gold_sales",
      "gold_supply"]
names_avail=[]
def name_check(name, jobs):
    return any(job in name for job in jobs)
for i in w.jobs.list():
    if(name_check(i.settings.name, jobs)):
        names_avail.append((i.settings.name, i.job_id))
print(names_avail)
for i in names_avail:
    job_settings = w.jobs.get(job_id=i[1]).settings
    job_settings.trigger.pause_status=PauseStatus.UNPAUSED
    w.jobs.reset(
        job_id=i[1],
        new_settings=job_settings
    )
```
