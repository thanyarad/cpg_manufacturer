```python
dbutils.widgets.text("catalog", "dev")
dbutils.widgets.text("schema", "00_landing")
dbutils.widgets.text("volume", "data")
```

```python
catalog=dbutils.widgets.get("catalog")
schema=dbutils.widgets.get("schema")
volume=dbutils.widgets.get("volume")
```

```python
root_path=f"/Volumes/{catalog}/{schema}/{volume}/"
input_file_paths = {
    "Consumer": "consumer",
    "Consumer Order": "consumer_orders/consumer_order",
    "Consumer Order Items": "consumer_orders/consumer_order_items",
    "Distributor": "distributor",
    "Distributor Invoice": "distributor_invoice/invoice",
    "Distributor Invoice Item": "distributor_invoice/invoice_item",
    "Distributor Sale Order": "distributor_sales_order/sale_order",
    "Distributor Sale Order Item": "distributor_sales_order/sale_order_item",
    "Inventory": "inventory",
    "Product": "product"
}

# testing paths
# for i in input_file_paths:
#   print(i," => ",root_path+input_file_paths[i])
```

```python
# Create a Unity Catalog volume using SQL
spark.sql(f"CREATE CATALOG IF NOT EXISTS {catalog}")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {catalog}.{schema}")
spark.sql(f"CREATE VOLUME IF NOT EXISTS {catalog}.{schema}.{volume}")
```

```python
import os

for entity_path in input_file_paths:
    full_path = os.path.join(root_path, input_file_paths[entity_path])
    if not os.path.exists(full_path):
        os.makedirs(full_path)
```
